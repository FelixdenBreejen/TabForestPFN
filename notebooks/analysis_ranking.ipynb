```python
from pathlib import Path
import xarray as xr
from tabularbench.results.ranking_table import make_ranking_table_, process_benchmark_results, process_sweep_results
from tabularbench.results.results_sweep import ResultsSweep
from tabularbench.utils.config_benchmark_sweep import ConfigBenchmarkSweep

outputdir = Path("outputs_done/tabzilla_tabsgfd_finetune/foundation-default-tabzilla_has_completed_runs")
outputdir1 = Path("outputs_done/tabzilla_tabpfn_original/tabpfn-default-tabzilla_has_completed_runs")
outputdir2 = Path("outputs_done/tabzilla_tabpfn_finetune/foundation-default-tabzilla_has_completed_runs")

cfg_forest_finetune = ConfigBenchmarkSweep.load(outputdir / "config_benchmark_sweep.yaml")
results_sweep_forest_finetune = ResultsSweep.load(outputdir / "results_sweep.nc")

cfg_tabpfn_original = ConfigBenchmarkSweep.load(outputdir1 / "config_benchmark_sweep.yaml")
results_sweep_tabpfn_original = ResultsSweep.load(outputdir1 / "results_sweep.nc")

cfg_tabpfn_finetune = ConfigBenchmarkSweep.load(outputdir2 / "config_benchmark_sweep.yaml")
results_sweep_tabpfn_finetune = ResultsSweep.load(outputdir2 / "results_sweep.nc")

ds_benchmark = process_benchmark_results(cfg_forest_finetune)
ds_sweep_forest_finetune = process_sweep_results(cfg_forest_finetune, results_sweep_forest_finetune)
ds_sweep_tabpfn_original = process_sweep_results(cfg_tabpfn_original, results_sweep_tabpfn_original)
ds_sweep_tabpfn_finetune = process_sweep_results(cfg_tabpfn_finetune, results_sweep_tabpfn_finetune)

ds_sweep_forest_finetune.coords['model_name'] = [cfg_forest_finetune.model_plot_name]
ds_sweep_tabpfn_original.coords['model_name'] = [cfg_tabpfn_original.model_plot_name]
ds_sweep_tabpfn_finetune.coords['model_name'] = [cfg_tabpfn_finetune.model_plot_name]

ds = xr.merge([ds_benchmark, ds_sweep_forest_finetune, ds_sweep_tabpfn_original, ds_sweep_tabpfn_finetune], combine_attrs='drop')


make_ranking_table_(cfg_forest_finetune, ds)
```
