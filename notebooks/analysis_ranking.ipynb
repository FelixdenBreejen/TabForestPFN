```python
from pathlib import Path
import xarray as xr
from tabularbench.results.ranking_table import make_ranking_table_, process_benchmark_results, process_sweep_results
from tabularbench.results.results_sweep import ResultsSweep
from tabularbench.utils.config_benchmark_sweep import ConfigBenchmarkSweep

def prepare_ds(path: Path):
    cfg = ConfigBenchmarkSweep.load(path / "config_benchmark_sweep.yaml")
    results_sweep = ResultsSweep.load(path / "results_sweep.nc")
    ds = process_sweep_results(cfg, results_sweep)
    ds.coords['model_name'] = [cfg.model_plot_name]
    return ds

cfg_forest_finetune = ConfigBenchmarkSweep.load(Path("outputs_done/tabzilla_tabsgfd_finetune/foundation-default-tabzilla_has_completed_runs") / "config_benchmark_sweep.yaml")
ds_benchmark = process_benchmark_results(cfg_forest_finetune)

ds_sweep_forest_finetune = prepare_ds(Path("outputs_done/tabzilla_tabsgfd_finetune/foundation-default-tabzilla_has_completed_runs"))
ds_sweep_tabpfn_original = prepare_ds(Path("outputs_done/tabzilla_tabpfn_original/tabpfn-default-tabzilla_has_completed_runs"))
ds_sweep_tabpfn_finetune = prepare_ds(Path("outputs_done/tabzilla_tabpfn_finetune/foundation-default-tabzilla_has_completed_runs"))
ds_sweep_forest_zeroshot = prepare_ds(Path("outputs_done/tabzilla_tabsgfd_zeroshot/foundation-default-tabzilla_has_completed_runs"))
ds_sweep_tabpfn_zeroshot = prepare_ds(Path("outputs_done/tabzilla_tabpfn_zeroshot/foundation-default-tabzilla_has_completed_runs"))

ds = xr.merge([
    ds_benchmark,
    ds_sweep_forest_finetune,
    ds_sweep_tabpfn_original,
    ds_sweep_tabpfn_finetune,
    # ds_sweep_forest_zeroshot,
    # ds_sweep_tabpfn_zeroshot
])


make_ranking_table_(cfg_forest_finetune, ds)
```
